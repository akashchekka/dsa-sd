```python
import heapq
```

```python
nums = [4, 1, 7, 3, 8, 5]

heapq.heapify(nums)
heapq.heappop(nums) # smallest element => 1
heapq.nsmallest(2, nums) # top k smallest [3, 4]

nums = [-k for k in nums]
heapq.heapify(nums) # Max heap
-heapq.heappop(nums) # largest element => 8
res = heapq.nsmallest(2, nums)

res = [-k for k in res]
res
```

### Heap Sort

```python
nums = [4, 2, 6, 1, 5, 3]

heapq.heapify(nums)
res = []

for i in range(len(nums)):
    num = heapq.heappop(nums)
    res.append(num)

res
```

### Kth largest element

```python
nums = [4, 1, 7, 3, 8, 5]

heap = []

for num in nums:
    heapq.heappush(heap, num)
    if len(heap) > 2:
        heapq.heappop(heap)
    
print(heap)
print(heapq.nlargest(3, nums)[-1])
```

### Top k frequent elements

```python
from collections import defaultdict, Counter
s = 'abbcccdddd'

# approach 1
res = []
defDict = defaultdict(list)
count = 1

for i in range(1, len(s)):
    if s[i] != s[i - 1]:
        defDict[count].append(s[i-1])
        count = 1
    else:
        count += 1
defDict[count].append(s[len(s) - 1])

keys = heapq.nlargest(2, defDict.keys())

for x in keys:
    res.extend(defDict[x])

print(res)

# approach 2
mostFrequent = [char for (char, _) in Counter(s).most_common(2)]
mostFrequent2 = [char for (char, _) in heapq.nlargest(2, Counter(s).items(), key = lambda x: x[1])]
print(mostFrequent)
print(mostFrequent2)
```
