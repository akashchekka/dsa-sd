Check if power of 2: n & (n-1) == 0 (only one bit set).
​
Set kth bit: n | (1 << k).

Clear kth bit: n & ~(1 << k).

Toggle kth bit: n ^ (1 << k).

Clear lowest set bit: n & (n-1).

Count set bits (popcount): Loop with n &= n-1 or Brian Kernighan.

Swap numbers: a ^= b; b ^= a; a ^= b.

### Swap without temp

```python
a = 2
b = 3

# a = a+b
# b = a-b
# a = a-b

a ^= b
b ^= a # b = b ^ (a ^ b)
a ^= b # a = (a ^ b) ^ a

print(a)
print(b)
```

### Count number of 1s

```python
a = 13
count = 0

while a:
    count += a&1
    a >>= 1

count
```

### Check if number is power of 2

```python
a = 4

'True' if a&(a-1) == 0 else 'False'
```

```python
# a << b = a * 2**b
# a >> b = a // 2**b

print(1<<3) # 1 * 2**3
print(3<<2) # 3 * 2**2
print(4>>1) # 4 * 2**1
```

### Add numbers without +

```python
MASK = 0xFFFFFFFF
MAX_INT = 2**31 - 1

a = -3
b = 4

# XOR (a ^ b) - Computes the sum without carrying
# AND + Left Shift ((a & b) << 1) - Computes the carry
# Masking simulates 32-bit integer overflow behavior

while b != 0:
    a, b = (a ^ b) & MASK, ((a & b) << 1) & MASK

print(a if a < MAX_INT else ~(a ^ MASK))
```

### Counting bits

```python
n = 5
ans = [0] * (n + 1)

print(list(range(n+1)))

for i in range(1, n+1):
    ans[i] = ans[i >> 1] + (i & 1)

ans    
```

### Reverse Bits

```python
n = 13 # 1101 => Reverse = 1011 (11)
res = 0
for i in range(32):
    bit = n & 1
    res = (res << 1) | bit
    n >>= 1
res
```
