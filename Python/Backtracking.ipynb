### Combination Sum I

- TC: O($2^t$*k) => t is target and k is length of curr appended to res
- SC: O(k*x) => x being auxiliary space

```python
def combination_sum(nums: list[int], target: int):
    n = len(nums)
    res = []

    def dfs(i, curr, total):
        if total == 0:
            res.append(curr.copy())
            return
        if total < 0: return
        
        for j in range(i, n):
            curr.append(nums[j])
            dfs(j, curr, total - nums[j])
            curr.pop()
    
    dfs(0, [], target)
    return res

combination_sum([2,5,6,9], 9)
```

### Combination Sum II

- TC: O(N.$2^N$) => $2^N$ subsets explored(for pick/non-pick decision), N for copy output 
- SC: O(N)

```python
def combination_sum_ii(nums: list[int], target: int):
    n = len(nums)
    nums.sort()
    res = []

    def dfs(i, curr, total):
        if total == 0:
            res.append(curr.copy())
            return
        if total < 0: return
        
        for j in range(i, n):
            if j > i and nums[j] == nums[j - 1]: continue

            curr.append(nums[j])
            dfs(j + 1, curr, total - nums[j])
            curr.pop()

    dfs(0, [], target)
    return res

combination_sum_ii([10, 1, 2, 7, 6, 1, 5], 8)
```

### Combination Sum III

```python
def combination_sum_iii(n: int, target: int):
    res = []

    def dfs(i, curr, total):
        if len(curr) == n:
            if total == 0:
                res.append(curr.copy())
            return
        if total < 0: return
        
        for j in range(i, 10):
            curr.append(j)
            dfs(j+1, curr, total - j)
            curr.pop()

    dfs(1, [], target)
    return res

combination_sum_iii(2, 7)
```

### Subsets

- subsets: O(N.$2^N$), N for [:] and $2^N$ for leaves(binary, include/exclue)
- subset_sum: O($2^N$), only leaves processing

```python
def subsets(nums: list[int]):
    res = []
    n = len(nums)

    def dfs(i, curr):
        if i == n:
            res.append(curr[:])
            return

        curr.append(nums[i])
        dfs(i + 1, curr)
        curr.pop()
        dfs(i + 1, curr)
    
    dfs(0, [])
    return res

def subset_sum(nums: list[int]):
    n = len(nums)
    res = []

    def dfs(i, total):
        if i == n:
            res.append(total)
            return
        
        total += nums[i]
        dfs(i + 1, total)
        total -= nums[i]
        dfs(i + 1, total)

    dfs(0, 0)
    return res

print(subsets([1, 2, 3]))
print(subset_sum([1, 2, 3]))
```

### Permutations

TC: O(N.N!)

```python
def permutations(nums: list[int]):
    n = len(nums)
    res = []

    def dfs(curr):
        if len(curr) == n:
            res.append(curr.copy())
            return
        
        for j in range(n):
            if nums[j] in curr: continue
            curr.append(nums[j])
            dfs(curr)
            curr.pop()
    
    dfs([])
    return res

def permutations_ii(nums: list[int]):
    n = len(nums)
    res = []

    def dfs(i):
        if i == n:
            res.append(nums.copy())
            return
        
        for j in range(i, n):
            nums[i], nums[j] = nums[j], nums[i]
            dfs(i + 1)
            nums[i], nums[j] = nums[j], nums[i]
    
    dfs(0)
    return res

permutations([1, 2, 3])
```

### N Queens

TC: O(N!) => c0 - N, c1 - (N-1) as we cant reuse row, ...
is_safe => O(N)

- With is_safe => O(N.N!)
- With set     => O(N!)

```python
def n_queens(n: int):
    res = []
    board = [['.'] * n for _ in range(n)]
    
    def is_safe(row, col):
        r, c = row, col
        
        # left top diag
        while r >=0 and c >= 0:
            if board[r][c] == 'Q': return False
            r -= 1
            c -= 1
        
        r, c = row, col
        # left bottom diag
        while r < n and c >= 0:
            if board[r][c] == 'Q': return False
            r += 1
            c -= 1

        r, c = row, col
        # left
        while c >= 0:
            if board[r][c] == 'Q': return False
            c -= 1

        return True

    # rows = set()
    # pos_diag = set()
    # neg_diag = set()
    
    def solve(col):
        nonlocal board
        if col == n:
            res.append([''.join(row) for row in board])
            return
        
        for row in range(n):
            if is_safe(row, col):
                board[row][col] = 'Q'
                solve(col + 1)
                board[row][col] = '.'

        # if (row not in rows) and (row + col not in pos_diag) and (row - col not in neg_diag):
        #         rows.add(row)
        #         pos_diag.add(row + col)
        #         neg_diag.add(row - col)
        #         board[row][col] = 'Q'

        #         solve(col + 1)

        #         rows.remove(row)
        #         pos_diag.remove(row + col)
        #         neg_diag.remove(row - col)
        #         board[row][col] = '.'

    solve(0)
    return res

n_queens(4)
```

### Solve Sudoku

- TC: sudoku_opt - O($9^E$) 
- SC: sudoku_opt - O(E) 

```python
def sudoku(board: list[list[str]]):
    def is_valid(r, c, s):
        for i in range(9):
            if board[i][c] == s: return False
            if board[r][i] == s: return False
            if board[3*(r//3) + i//3][3*(c//3) + i%3] == s: return False
        return True

    def solve(board: list[list[str]]):
        for i in range(9):
            for j in range(9):
                if board[i][j] == '.':
                    for c in range(1, 10):
                        if is_valid(i, j, str(c)):
                            board[i][j] = str(c)
                            if solve(board): return True
                            else: board[i][j] = '.'
                    return False
        return True
    
    solve(board)
    for row in board:
        print(row)

def sudoku_opt(board: list[list[str]]):
    rows = [set() for _ in range(9)]
    cols = [set() for _ in range(9)]
    boxes = [set() for _ in range(9)]
    empties = []

    for r in range(9):
        for c in range(9):
            cell = board[r][c]
            if cell == '.':
                empties.append((r, c))
            else:
                rows[r].add(cell)
                cols[c].add(cell)
                boxes[3*(r//3) + (c//3)].add(cell)

    def solve(k: int):
        if len(empties) == k: return True

        r, c = empties[k]
        b = 3 * (r//3) + (c//3)
        for s in '123456789':
            if s in rows[r] or s in cols[c] or s in boxes[b]: continue
            board[r][c] = s
            rows[r].add(s); cols[c].add(s); boxes[b].add(s)
            if solve(k + 1): return True
            board[r][c] = '.'
            rows[r].remove(s); cols[c].remove(s); boxes[b].remove(s)
        return False
    
    solve(0)
    for row in board:
        print(row)

board = [["5","3",".",".","7",".",".",".","."],["6",".",".","1","9","5",".",".","."],[".","9","8",".",".",".",".","6","."],["8",".",".",".","6",".",".",".","3"],["4",".",".","8",".","3",".",".","1"],["7",".",".",".","2",".",".",".","6"],[".","6",".",".",".",".","2","8","."],[".",".",".","4","1","9",".",".","5"],[".",".",".",".","8",".",".","7","9"]]
board = [[".",".",".",".",".",".",".",".","."],[".","9",".",".","1",".",".","3","."],[".",".","6",".","2",".","7",".","."],[".",".",".","3",".","4",".",".","."],["2","1",".",".",".",".",".","9","8"],[".",".",".",".",".",".",".",".","."],[".",".","2","5",".","6","4",".","."],[".","8",".",".",".",".",".","1","."],[".",".",".",".",".",".",".",".","."]]
sudoku_opt(board)
```
