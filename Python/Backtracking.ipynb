### Combination Sum I

```python
def combination_sum(nums: list[int], target: int):
    n = len(nums)
    res = []

    def dfs(i, curr, total):
        if total == 0:
            res.append(curr.copy())
            return
        
        for j in range(i, n):
            if nums[j] > total: continue

            curr.append(nums[j])
            dfs(j, curr, total - nums[j])
            curr.pop()
    
    dfs(0, [], target)
    return res

combination_sum([2,5,6,9], 9)
```

### Combination Sum II

```python
def combination_sum_ii(nums: list[int], target: int):
    n = len(nums)
    nums.sort()
    res = []

    def dfs(i, curr, total):
        if total == 0:
            res.append(curr.copy())
            return
        
        for j in range(i, n):
            if j > i and nums[j] == nums[j - 1]: continue
            if nums[j] > total: break

            curr.append(nums[j])
            dfs(j + 1, curr, total - nums[j])
            curr.pop()

    dfs(0, [], target)
    return res

combination_sum_ii([10, 1, 2, 7, 6, 1, 5], 8)
```

### Combination Sum III

```python
def combination_sum_iii(n: int, target: int):
    res = []

    def dfs(i, curr, total):
        if len(curr) == n:
            if total == 0:
                res.append(curr.copy())
            return
        
        for j in range(i, 10):
            if j > total: break

            curr.append(j)
            dfs(j+1, curr, total - j)
            curr.pop()

    dfs(1, [], target)
    return res

combination_sum_iii(2, 7)
```

### Subsets

```python
def subsets(nums: list[int]):
    res = []
    n = len(nums)

    def dfs(i, curr):
        if i == n:
            res.append(curr.copy())
            return

        curr.append(nums[i])
        dfs(i + 1, curr)
        curr.pop()
        dfs(i + 1, curr)
    
    dfs(0, [])
    return res

def subset_sum(nums: list[int]):
    n = len(nums)
    res = []

    def dfs(i, total):
        if i == n:
            res.append(total)
            return
        
        total += nums[i]
        dfs(i + 1, total)
        total -= nums[i]
        dfs(i + 1, total)

    dfs(0, 0)
    return res

print(subset_sum([1, 2, 3]))
print(subsets([1, 2, 3]))
```

### Permutations

```python
from bisect import bisect_right

def permutations(nums: list[int]):
    n = len(nums)
    res = []

    def dfs(curr, hset: set):
        if len(curr) == n:
            res.append(curr.copy())
            return
        
        for j in range(n):
            if j not in hset:
                hset.add(j)
                curr.append(nums[j])
                dfs(curr, hset)
                hset.remove(j)
                curr.pop()
    
    dfs([], set())
    return res

def permutations_ii(nums: list[int]):
    n = len(nums)
    res = []
    curr = None

    def dfs(i):
        nonlocal curr
        if i == n:
            res.append(curr.copy())
            return
        
        for j in range(i, n):
            nums[i], nums[j] = nums[j], nums[i]
            curr = nums
            dfs(i + 1)
            nums[i], nums[j] = nums[j], nums[i]
    
    dfs(0)
    return res

ans = permutations_ii([1, 2, 3])
print(ans)
index = bisect_right(ans, [2, 1, 3])
print(index)
```
