```python
class Trie:
    def __init__(self):
        self.isEnd = False
        self.children = dict()

    def add_word(self, word):
        trie = self
        for char in word:
            if char not in trie.children:
                trie.children[char] = Trie()
            trie = trie.children[char]
        
        trie.isEnd = True

    def search(self, word):
        trie = self
        for char in word:
            if char not in trie.children:
                return False
            trie = trie.children[char]
        
        return trie.isEnd
    
    def autocomplete_util(self, trie, suffix, result):
        if trie.isEnd:
            result.append(suffix)

        for char in trie.children:
            self.autocomplete_util(trie.children[char], suffix + char, result)
    
    def autocomplete(self, prefix):
        trie = self
        for char in prefix:
            if char not in trie.children:
                return []
            trie = trie.children[char]
        
        result = []
        self.autocomplete_util(trie, prefix, result)
        return result
```

```python
trie = Trie()

trie.add_word('bun')
trie.add_word('bundle')
trie.add_word('burn')

trie.autocomplete('bu')
```
