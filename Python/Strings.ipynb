### Luhn's Algorithm - Credit Card number check

```python
string = '36101025079892'

str_array = [int(string[i]) for i in reversed(range(len(string)))]

sum = 0
for i, x in enumerate(str_array):
    if i & 1:
        doubled = (2 * x)
        doubled_mod_nine = doubled - 9 if doubled > 9 else doubled
        sum += doubled_mod_nine
    else:
        sum += x

sum % 10 == 0
```

### Minimum Window Substring

```python
# ADOBECODEBANC <==> BAC

from collections import defaultdict

def min_window_substring(s:str , t:str):
    window = defaultdict(int)
    need = defaultdict(int)

    for c in t:
        need[c] += 1

    valid = 0
    start, length = 0, float('inf')
    l = r = 0

    while r < len(s):
        current = s[r]
        r += 1

        # Update the valid if the count of char is equal to count in need
        if current in need:
            window[current] += 1
            if window[current] == need[current]:
                valid += 1

        while valid == len(need):
            if r - l < length:
                start = l
                length = r - l # Update length once valid = len(need) 

            l2 = s[l]
            l += 1

            # Shrink window from left
            if l2 in need:
                if window[l2] == need[l2]:
                    valid -= 1
                window[l2] -= 1
    
    if length == float('inf'): return ""

    return s[start:start+length]

min_window_substring("ADOBECODEBANC", "ABC")
```

### String to Int

```python
s = '  -042'

def str_to_int(s: str):
    s = s.lstrip()

    if not s:
        return 0
    
    i = 0
    sign = -1 if s[i] == '-' else 1
    i += 1 if s[i] in ['-', '+'] else 0

    res = 0
    while i < len(s) and s[i].isdigit():
        res = res * 10 + int(s[i])
        i += 1

    res *= sign

    return max(-2**31, min(res, 2**31 - 1))

str_to_int(s)
```
