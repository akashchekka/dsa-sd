### Luhn's Algorithm - Credit Card number check

```python
string = '36101025079892'

str_array = [int(string[i]) for i in reversed(range(len(string)))]

sum = 0
for i, x in enumerate(str_array):
    if i & 1:
        doubled = (2 * x)
        doubled_mod_nine = doubled - 9 if doubled > 9 else doubled
        sum += doubled_mod_nine
    else:
        sum += x

sum % 10 == 0
```
