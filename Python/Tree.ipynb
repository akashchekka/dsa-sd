```python
class Node:
    def __init__(self, data: int):
        self.data = data
        self.left = None
        self.right = None
```

```python
import collections

class BST:
    def __init__(self):
        self.head = None

    ### Insert Nodes to tree
    def insert(self, data: int):
        self.head = self.add_node(self.head, data)

    def add_node(self, head: Node, data: int):
        if head is None:
            return Node(data)
        
        if data < head.data:
            head.left = self.add_node(head.left, data)
        elif data > head.data:
            head.right = self.add_node(head.right, data)

        return head
    
    ### Print tree
    def print_tree(self, head: Node):
        if head is None:
            return None
        
        self.print_tree(head.left)
        print(head.data, end=' ')
        self.print_tree(head.right)

    def display(self):
        self.print_tree(self.head)

    ### BFS
    def bfs(self, head: Node):
        queue = collections.deque()
        queue.append(head)

        while queue:
            first_node = queue.popleft()
            print(first_node.data)
            if first_node.left:
                queue.append(first_node.left)
            if first_node.right:
                queue.append(first_node.right)

    ### right view
    def right_view(self, head: Node):
        q = collections.deque()
        q.append(head)

        while q:
            q_len = len(q)
            rightNode = None
            for i in range(q_len):
                node = q.popleft()
                if node:
                    rightNode = node
                    if node.left:
                        q.append(node.left)
                    if node.right:
                        q.append(node.right)
            print(rightNode.data, end=' ')

    ### Depth of tree:
    def max_depth(self, head: Node):
        if head is None:
            return -1
        
        return max(self.max_depth(head.left), self.max_depth(head.right)) + 1

    ### Print level k
    def print_level_util(self, head: Node, c: int, k: int):
        if head is None:
            return
        
        if c == k:
            print(head.data, end=' ')
            self.levelCount += 1
            return

        self.print_level_util(head.left, c+1, k)
        self.print_level_util(head.right, c+1, k)

    def print_level(self, k: int):
        c = 0
        self.levelCount = 0
        self.print_level_util(self.head, c, k)
        print("\ncount: " + str(self.levelCount))

    

    def delete_key(self, key):
        def delete_key_util(self, node, key):
            if not node:
                return None
            
            temp = node

            if key < temp.data:
                temp.left = self.delete_key_util(temp.left, key)
            elif key > temp.data:
                temp.right = self.delete_key_util(temp.right, key)
            else:
                if temp.right is None:
                    return temp.left
                if temp.left is None:
                    return temp.right
                
                leafNode = temp.right
                while leafNode.left:
                    leafNode = leafNode.left
                
                temp.data = leafNode.data
                temp.right = self.delete_key_util(temp.right, leafNode.data)

            return temp
        
        return delete_key_util(self.head, key)

    def lowest_common_ancestor(self, p: Node, q: Node):
        node = self.head

        while node:
            if p.data > node.data and q.data > node.data:
                node = node.right
            elif p.data < node.data and q.data < node.data:
                node = node.left
            else:
                return node.data

    def kthSmallest(self, k):
        node = self.head

        c = 0
        res = 0

        def traverse(node):
            nonlocal c, res

            if not node:
                return None
            
            traverse(node.left)

            c += 1
            if c == k:
                res = node.data
                return

            traverse(node.right)

        traverse(node)
        print(res)

    def isSubtree(self, root, subRoot) -> bool:
        if not root:
            return False

        if not subRoot:
            return True

        def verify(root1, root2):
            if not root1 and not root2:
                return True
            
            if not root1 or not root2:
                return False

            if root1.val != root2.val:
                return False

            return (True and 
                    verify(root1.left, root2.left) and 
                    verify(root1.right, root2.right))

        def inorder(root):
            if not root:
                return False

            if root.val == subRoot.val:
                if verify(root, subRoot):
                    return True

            return inorder(root.left) or inorder(root.right)

        return inorder(root)

    def isValidBST(self, root: Node) -> bool:
        from collections import deque
        
        if not root:
            return True

        q = deque([(root, float('-inf'), float('inf'))])

        while q:
            curr, l, r = q.popleft()

            if not (l < curr.val < r):
                return False

            if curr.left:
                q.append((curr.left, l, curr.val))
            if curr.right:
                q.append((curr.right, curr.val, r))

        return True
    
    def serialize_and_deserialize(self, root: Node):
        def serialize(root: Node):
            res = []

            def dfs(node):
                if not node:
                    res.append('N')
                    return None
                
                res.append(str(root.val))
                dfs(root.left)
                dfs(root.right)
            
            dfs(root)
            return ','.join(res)
        
        def deserialize(data: str):
            data = data.split(',')
            i = 0

            def dfs():
                nonlocal i
                if data[i] == 'N':
                    i += 1
                    return None
                
                node = Node(int(data[i]))

                node.left = dfs()
                node.right = dfs()

                return node
            
            return dfs()
        
    def diameter(self, root:Node):
        res = 0

        def dfs(node):
            nonlocal res

            if not node:
                return 0
            
            left = dfs(node.left)
            right = dfs(node.right)
            
            res = max(res, left + right)

            return 1 + max(left, right)
        
        dfs(root)

        return res
```

```python
## Level Order Printing

tree = BST()

nodes = [16, 8, 24, 4, 12, 20, 28]

for node in nodes:
    tree.insert(node)

head = tree.head

q = collections.deque()
q.append(head)

res = []
while q:
    qlen = len(q)
    subnodes = []
    for i in range(qlen):
        temp = q.popleft()
        subnodes.append(temp.data)
        if temp.left is not None:
            q.append(temp.left)
        if temp.right is not None:
            q.append(temp.right)
    
    res.append(subnodes)

res
```
