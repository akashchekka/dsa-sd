```python
from abc import ABC, abstractmethod
from datetime import datetime, timezone
```

```python
class IClock(ABC):
    @abstractmethod
    def now(self) -> datetime: ...
```

```python
class SystemClock(IClock):
    def now(self) -> datetime:
        return datetime.now(timezone.utc)
```

```python
from typing import TypeVar, Generic, Hashable, Optional
from abc import ABC, abstractmethod

K = TypeVar("K", bound=Hashable)
V = TypeVar("V")

class ICache(ABC, Generic[K, V]):
    @abstractmethod
    def get(self, key: K) -> Optional[V]: ...
    @abstractmethod
    def put(self, key: K, value: V) -> None: ...
    @abstractmethod
    @property
    def count(self) -> int: ...
```

```python
import threading
import time
import random
from typing import Generic, TypeVar, List

T = TypeVar('T')

class BoundedBlockingQueue(Generic[T]):
    def __init__(self, capacity: int):
        self._capacity = capacity
        self._queue: List[T] =[]
        
        # A Condition variable is inherently bound to a Lock
        self._condition = threading.Condition()

    def enqueue(self, item: T) -> None:
        with self._condition:
            # Using 'while' instead of 'if' prevents spurious wakeups
            while len(self._queue) >= self._capacity:
                print("Queue FULL. Producer waiting...")
                self._condition.wait()
                
            self._queue.append(item)
            print(f"Produced: {item}. Queue size: {len(self._queue)}")
            
            # Notify consumers that data is available
            self._condition.notify_all()

    def dequeue(self) -> T:
        with self._condition:
            while len(self._queue) == 0:
                print("Queue EMPTY. Consumer waiting...")
                self._condition.wait()
                
            item = self._queue.pop(0)
            print(f"Consumed: {item}. Queue size: {len(self._queue)}")
            
            # Notify producers that space is available
            self._condition.notify_all()
            return item
        
queue: BoundedBlockingQueue[int] = BoundedBlockingQueue(capacity=3)

def producer(name: str):
    for _ in range(5):
        item = random.randint(1, 100)
        print(f"[{name}] trying to produce {item}")
        queue.enqueue(item)
        time.sleep(random.uniform(0.05, 0.2))

def consumer(name: str):
    for _ in range(5):
        print(f"[{name}] trying to consume")
        item = queue.dequeue()
        print(f"[{name}] got {item}")
        time.sleep(random.uniform(0.05, 0.2))

threads = [
    threading.Thread(target=producer, args=("P1",)),
    threading.Thread(target=consumer, args=("C1",))
]

for t in threads:
    t.start()

for t in threads: t.join()
print("Done. Final queue size:", len(queue._queue))
```
