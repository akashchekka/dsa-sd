### Fibonacci

```python
# Memoization
def fib(n):
    memo = {}

    def dp(n):
        if n <= 2: return 1
        if n not in memo:
            memo[n] = dp(n - 1) + dp(n - 2)
        return memo[n]
    
    return dp(n)

# Tabulation
def fib_tab(n):
    arr = [0] * (n + 1)
    arr[1] = 1

    for i in range(n):
        arr[i + 1] += arr[i]
        if i + 2 <= n:
            arr[i + 2] += arr[i]

    print(arr[n]) 
print(fib(50))
fib_tab(50)
```

### Grid Traveller

TC: O(m x n)

SC: O(m x n) memo table + O(m + n) recursion stack

dp(m, n) = dp(m - 1, n) + dp(m, n - 1)

```python
def grid_traveller(m, n):
    memo = {}

    def dp(m, n):
        if m == 1 and n == 1: return 1
        if m == 0 or n == 0: return 0

        pos = (m, n)
        if pos not in memo:
            memo[pos] = dp(m - 1, n) + dp(m, n - 1)
        return memo[pos]

    return dp(m, n)

grid_traveller(3, 7)
```

### CanSum

```python
def can_sum(target, nums):
    memo = {}

    def dp(target):
        if target == 0: return True
        if target < 0: return False

        for num in nums:
            rem = target - num
            if rem not in memo:
                memo[rem] = dp(rem)
            if memo[rem]: return True
            
        return False
    
    return dp(target)

def can_sum2(target, nums):
    tab = [False] * (target + 1)
    tab[0] = True

    for i in range(target + 1):
        for j in nums:
            if tab[i] and i + j <= target:
                tab[i + j] = True
    return tab[target]
can_sum2(7, [3, 2])
```

### Return a combination that add up to target if exists

```python
# Memoization
def howSum(target, nums):
    memo = {}

    def dp(target):
        if target in memo: return memo[target]
        if target == 0: return []
        if target < 0: return None

        for num in nums:
            rem = target - num
            res = dp(rem)
            if res is not None:
                memo[target] = res + [num]
                return memo[target]
        
        memo[target] = None
        return None
    
    return dp(target)

# Tabulation
# m = len(target)
# n = len(nums)
# TC: O(m^2 * n)
# SC: O(m^2)
def howSum2(target, nums):
    arr = [None] * (target + 1)
    arr[0] = []

    for i in range(target + 1):
        for j in nums:
            if arr[i] is not None and (i + j) <= target:
                arr[i + j] = [j] + arr[i]
    return arr[target]

print(howSum2(7, [2, 3, 7]))
```

### Best Sum

```python
def best_sum(target, nums):
    memo = {}

    def dp(target):
        if target in memo: return memo[target]
        if target == 0: return []
        if target < 0: return None

        shortestCombination = None

        for num in nums:
            rem = target - num
            res = dp(rem)
            if res is not None:
                combination = res + [num]
                if not shortestCombination or combination > shortestCombination:
                    shortestCombination = combination
        
        memo[target] = shortestCombination
        return shortestCombination
    
    return dp(target)

def best_sum2(target, nums):
    arr = [None] * (target + 1)
    arr[0] = []

    for i in range(target + 1):
        for j in nums:
            if arr[i] is not None and i + j <= target:
                newElement = arr[i] + [j]
                if arr[i + j] is None or len(newElement) < len(arr[i + j]):
                    arr[i + j] = newElement
    
    return arr[target]

best_sum2(7, [2, 3, 7])
```

```python
def canConstruct(target: str, wordList: list[str]):
    memo = {}

    def dp(target):
        if target in memo: return memo[target]
        if target == '': return True

        for word in wordList:
            if target.startswith(word):
                suffix = target.removeprefix(word)
                if dp(suffix):
                    memo[target] = True
                    return True
                
        memo[target] = False
        return False
    
    return dp(target)

def canConstruct2(target: str, wordList: list[str]):
    n = len(target)
    arr = [False] * (n + 1)
    arr[0] = True

    for i in range(n + 1):
        for j in wordList:
            if arr[i] and i + len(j) <= n and target.startswith(j, i):
                arr[i + len(j)] = True
    
    return arr[n]

# canConstruct2('abcdef', ['ab', 'abc', 'cd', 'def', 'abcd'])
canConstruct2('abcdef', ['ab', 'abc'])
```

### Count total number of combinations from wordlist are possible to construct target

```python
def countConstruct(target, wordlist):
    memo = {}

    def dp(target):
        if target in memo: return memo[target]
        if target == '': return 1

        count = 0
        for word in wordlist:
            if target.startswith(word):
                suffix = target.removeprefix(word)
                count += dp(suffix)

        memo[target] = count
        return count
    
    return dp(target)

def countConstruct2(target: str, wordList: list[str]):
    n = len(target)
    arr = [0] * (n + 1)
    arr[0] = 1

    for i in range(n + 1):
        for j in wordList:
            if arr[i] != 0 and i + len(j) <= n and target.startswith(j, i):
                arr[i + len(j)] += arr[i]

    return arr[n]

countConstruct2('abcdef', ['ab', 'abc', 'cd', 'def', 'ef'])
# countConstruct2('abcdef', ['ab', 'abc'])
```

### All combinations of wordlist to build target

```python
def allConstruct(target: str, wordlist: list[str]):
    memo = {}

    def dp(target):
        if target in memo: return memo[target]
        if target == '': return [[]]

        res = []
        for word in wordlist:
            if target.startswith(word):
                suffix = target.removeprefix(word)
                suffixWays = dp(suffix)
                targetWays = [[word, *way] for way in suffixWays]
                res += [*targetWays]
        
        memo[target] = res
        return res

    return dp(target)

def allConstruct2(target: str, wordlist: list[str]):
    n = len(target)
    arr = [[] for _ in range(n + 1)]
    arr[0] = [[]]

    for i in range(n + 1):
        for j in wordlist:
            if arr[i] is not None and i + len(j) <= n and target.startswith(j, i):
                newCombination = [[*way, j] for way in arr[i]]
                arr[i + len(j)] += newCombination
    
    return arr[n]

allConstruct2('abcdef', ['ab', 'abc', 'cd', 'def', 'ef'])
# allConstruct('aaaaaaaaaaaaaaaaaaaaaaaaaaaz', ['a', 'aa', 'aaaaa', 'aaaaa', 'aaa'])
```

### Coin Change I

- Minimum number of coins to make amount
- arr[0] = 0 => 0 coins are needed to make 0
- arr[i] = None

```python
def change(amount: int, coins: list[int]):
    arr = [None] * (amount + 1)
    arr[0] = 0

    for i in range(amount + 1):
        for j in coins:
            if arr[i] is not None and i + j <= amount and (arr[i + j] is None or arr[i] + 1 < arr[i + j]):
                arr[i + j] = arr[i] + 1

    return arr[amount] if arr[amount] is not None else -1

change(4, [1, 2, 3])
```

### Coin Change II

- Number of distinct combinations to make ammount
- arr[0] = 1 => 1 way to make 0, empty combination
- arr[i] = 0 => 0 ways till we find one

```python
def change(amount: int, coins: list[int]) -> int:
    arr = [0] * (amount + 1)
    arr[0] = 1

    for j in coins:
        for i in range(amount + 1):
            if i + j <= amount:
                arr[i + j] += arr[i]

    return arr[amount]

change(4, [1, 2, 3])
```

### Climb Stairs

- How many distinct ways can you reach the top
- Permutations

```python
def climb_stairs(n):
    nums = [1, 2]
    arr = [0] * (n + 1)
    arr[0] = 1

    for i in range(len(arr)):
        for j in nums:
            if i + j <= n:
                arr[i + j] += arr[i]

    return arr[n]

climb_stairs(3)
```

### Climb Stairs II

- How many unique combinations to climb to top? (1, 2) and (2, 1) both are same
- Combinations

```python
def climb_stairs(n):
    nums = [1, 2]
    arr = [0] * (n + 1)
    arr[0] = 1

    for j in nums:
        for i in range(len(arr)):
            if i + j <= n:
                arr[i + j] += arr[i]

    return arr[n]

climb_stairs(3)
```
