### Fibonacci

```python
def fib(n):
    memo = {}

    def dp(n):
        if n <= 2: return 1
        if n not in memo:
            memo[n] = dp(n - 1) + dp(n - 2)
        return memo[n]
    
    return dp(n)

fib(50)
```

### Grid Traveller

```python
def grid_traveller(m, n):
    memo = {}

    def dp(m, n):
        if m == 1 and n == 1: return 1
        if m == 0 or n == 0: return 0

        pos = (m, n)
        if pos not in memo:
            memo[pos] = dp(m - 1, n) + dp(m, n - 1)
        return memo[pos]

    return dp(m, n)

grid_traveller(3, 7)
```
