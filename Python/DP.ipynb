### Fibonacci

```python
# Memoization
def fib(n):
    memo = {}

    def dp(n):
        if n <= 2: return 1
        if n not in memo:
            memo[n] = dp(n - 1) + dp(n - 2)
        return memo[n]
    
    return dp(n)

# Tabulation
def fib_tab(n):
    arr = [0] * (n + 1)
    arr[1] = 1

    for i in range(n):
        arr[i + 1] += arr[i]
        if i + 2 <= n:
            arr[i + 2] += arr[i]

    print(arr[n]) 
print(fib(50))
fib_tab(50)
```

### Grid Traveller

TC: O(m x n)

SC: O(m x n) memo table + O(m + n) recursion stack

dp(m, n) = dp(m - 1, n) + dp(m, n - 1)

```python
def grid_traveller(m, n):
    memo = {}

    def dp(m, n):
        if m == 1 and n == 1: return 1
        if m == 0 or n == 0: return 0

        pos = (m, n)
        if pos not in memo:
            memo[pos] = dp(m - 1, n) + dp(m, n - 1)
        return memo[pos]

    return dp(m, n)

grid_traveller(3, 7)
```

### CanSum

```python
def can_sum(target, nums):
    memo = {}

    def dp(target):
        if target == 0: return True
        if target < 0: return False

        for num in nums:
            rem = target - num
            if rem not in memo:
                memo[rem] = dp(rem)
            if memo[rem]: return True
            
        return False
    
    return dp(target)

def can_sum2(target, nums):
    tab = [False] * (target + 1)
    tab[0] = True

    for i in range(target + 1):
        for j in nums:
            if tab[i] and i + j <= target:
                tab[i + j] = True
    return tab[target]
can_sum2(7, [3, 2])
```

### Return a combination that add up to target if exists

```python
# Memoization
def howSum(target, nums):
    memo = {}

    def dp(target):
        if target in memo: return memo[target]
        if target == 0: return []
        if target < 0: return None

        for num in nums:
            rem = target - num
            res = dp(rem)
            if res is not None:
                memo[target] = res + [num]
                return memo[target]
        
        memo[target] = None
        return None
    
    return dp(target)

# Tabulation
# m = len(target)
# n = len(nums)
# TC: O(m^2 * n)
# SC: O(m^2)
def howSum2(target, nums):
    arr = [None] * (target + 1)
    arr[0] = []

    for i in range(target + 1):
        for j in nums:
            if arr[i] is not None and (i + j) <= target:
                arr[i + j] = [j] + arr[i]
    return arr[target]

print(howSum2(7, [2, 3, 7]))
```

### Best Sum

```python
def best_sum(target, nums):
    memo = {}

    def dp(target):
        if target in memo: return memo[target]
        if target == 0: return []
        if target < 0: return None

        shortestCombination = None

        for num in nums:
            rem = target - num
            res = dp(rem)
            if res is not None:
                combination = res + [num]
                if not shortestCombination or combination > shortestCombination:
                    shortestCombination = combination
        
        memo[target] = shortestCombination
        return shortestCombination
    
    return dp(target)

def best_sum2(target, nums):
    arr = [None] * (target + 1)
    arr[0] = []

    for i in range(target + 1):
        for j in nums:
            if arr[i] is not None and i + j <= target:
                newElement = arr[i] + [j]
                if arr[i + j] is None or len(newElement) < len(arr[i + j]):
                    arr[i + j] = newElement
    
    return arr[target]

best_sum2(7, [2, 3, 7])
```

```python
def canConstruct(target: str, wordList: list[str]):
    memo = {}

    def dp(target):
        if target in memo: return memo[target]
        if target == '': return True

        for word in wordList:
            if target.startswith(word):
                suffix = target.removeprefix(word)
                if dp(suffix):
                    memo[target] = True
                    return True
                
        memo[target] = False
        return False
    
    return dp(target)

def canConstruct2(target: str, wordList: list[str]):
    n = len(target)
    dp = [False] * (n + 1)
    dp[0] = True

    for i in range(n + 1):
        if not dp[i]: continue
        for word in wordList:
            if i + len(word) <= n and target.startswith(word, i):
                dp[i + len(word)] = True

    return dp[n]

# canConstruct2('abcdef', ['ab', 'abc', 'cd', 'def', 'abcd'])
canConstruct2('abcdef', ['ab', 'abc'])
```

### Count total number of combinations from wordlist are possible to construct target

```python
def countConstruct(target, wordlist):
    memo = {}

    def dp(target):
        if target in memo: return memo[target]
        if target == '': return 1

        count = 0
        for word in wordlist:
            if target.startswith(word):
                suffix = target.removeprefix(word)
                count += dp(suffix)

        memo[target] = count
        return count
    
    return dp(target)

def countConstruct2(target: str, wordList: list[str]):
    n = len(target)
    arr = [0] * (n + 1)
    arr[0] = 1

    for i in range(n + 1):
        for j in wordList:
            if arr[i] != 0 and i + len(j) <= n and target.startswith(j, i):
                arr[i + len(j)] += arr[i]

    return arr[n]

countConstruct2('abcdef', ['ab', 'abc', 'cd', 'def', 'ef'])
# countConstruct2('abcdef', ['ab', 'abc'])
```

### All combinations of wordlist to build target

```python
def allConstruct(target: str, wordlist: list[str]):
    memo = {}

    def dp(target):
        if target in memo: return memo[target]
        if target == '': return [[]]

        res = []
        for word in wordlist:
            if target.startswith(word):
                suffix = target.removeprefix(word)
                suffixWays = dp(suffix)
                targetWays = [[word, *way] for way in suffixWays]
                res += [*targetWays]
        
        memo[target] = res
        return res

    return dp(target)

def allConstruct2(target: str, wordlist: list[str]):
    n = len(target)
    arr = [[] for _ in range(n + 1)]
    arr[0] = [[]]

    for i in range(n + 1):
        for j in wordlist:
            if arr[i] is not None and i + len(j) <= n and target.startswith(j, i):
                newCombination = [[*way, j] for way in arr[i]]
                arr[i + len(j)] += newCombination
    
    return arr[n]

allConstruct2('abcdef', ['ab', 'abc', 'cd', 'def', 'ef'])
# allConstruct('aaaaaaaaaaaaaaaaaaaaaaaaaaaz', ['a', 'aa', 'aaaaa', 'aaaaa', 'aaa'])
```

### Coin Change I

- Minimum number of coins to make amount
- arr[0] = 0 => 0 coins are needed to make 0
- arr[i] = None

```python
def change(amount: int, coins: list[int]):
    arr = [None] * (amount + 1)
    arr[0] = 0

    for i in range(amount + 1):
        for j in coins:
            if arr[i] is not None and i + j <= amount and (arr[i + j] is None or arr[i] + 1 < arr[i + j]):
                arr[i + j] = arr[i] + 1

    return arr[amount] if arr[amount] is not None else -1

change(4, [1, 2, 3])
```

### Coin Change II

- Number of distinct combinations to make ammount
- arr[0] = 1 => 1 way to make 0, empty combination
- arr[i] = 0 => 0 ways till we find one

```python
def change(amount: int, coins: list[int]) -> int:
    arr = [0] * (amount + 1)
    arr[0] = 1

    for j in coins:
        for i in range(amount + 1):
            if i + j <= amount:
                arr[i + j] += arr[i]

    return arr[amount]

change(4, [1, 2, 3])
```

### Climb Stairs

- How many distinct ways can you reach the top
- Permutations

```python
def climb_stairs(n):
    nums = [1, 2]
    arr = [0] * (n + 1)
    arr[0] = 1

    for i in range(len(arr)):
        for j in nums:
            if i + j <= n:
                arr[i + j] += arr[i]

    return arr[n]

def climb_stairs2(n):
    memo = [-1] * (n + 1)

    def dp(n):
        if memo[n] != -1: return memo[n]
        if n == 0: return 1

        ways = 0
        for i in range(2):
            if n - (i+1) >= 0:
                ways += dp(n - (i + 1))

        memo[n] = ways
        return ways
    
    return dp(n)

climb_stairs2(3)
```

### Frog Jump

```python
def frog_jump(n: int, heights: list[int]):
    memo = [-1] * n

    def dp(index):
        if index == 0: return 0
        if memo[index] != -1: return memo[index]

        left = dp(index - 1) + abs(heights[index] - heights[index - 1])
        right = float('inf')
        if index >= 2:
            right = dp(index - 2) + abs(heights[index] - heights[index - 2])

        memo[index] = min(left, right)
        return memo[index]
    
    return dp(n - 1)

def frog_jump2(n: int, heights: list[int], k):
    memo = [-1] * n

    def dp(index):
        if index == 0: return 0
        if memo[index] != -1: return memo[index]

        min_energy = float('inf')
        for i in range(k):
            if index - (i + 1) >= 0:
                energy_i = dp(index - (i + 1)) + abs(heights[index] - heights[index - (i + 1)])
                min_energy = min(min_energy, energy_i)

        memo[index] = min_energy
        return memo[index]
    
    return dp(n - 1)

def frog_jump_tab(n: int, heights: list[int], k):
    dp = [0] * n
    dp[0] = 0
    
    for index in range(1, n):
        min_energy = float('inf')
        for i in range(1, k+1):
            if index - i >= 0:
                energy_i = dp[index - i] + abs(heights[index] - heights[index - i])
                min_energy = min(min_energy, energy_i)

        dp[index] = min_energy
    
    return dp[n - 1]

frog_jump_tab(6, [30, 10, 60, 10, 60, 50], 2)
```

### Min Cost Climbing Stairs

```python
def minCostClimbingStairs(cost: list[int]) -> int:
    n = len(cost)
    dp = [0] * (n + 1)
    
    for i in range(2, n + 1):
        cost_1 = dp[i - 1] + cost[i - 1]
        cost_2 = dp[i - 2] + cost[i - 2]

        dp[i] = min(cost_1, cost_2)
    
    return dp[n]

def minCostClimbingStairs_spaceopt(cost: list[int]):
    n = len(cost)
    prev = prev2 = 0

    for i in range(2, n+1):
        cost1 = prev + cost[i - 1]
        cost2 = prev2 + cost[i - 2]

        curr = min(cost1, cost2)

        prev2 = prev
        prev = curr

    return prev

minCostClimbingStairs_spaceopt([1,2,1,2,1,1,1])
```

### Max Sum of Non Adjacent Elements (House Robber - I)

```python
def max_sum_memoization(nums: list[int]):
    n = len(nums)
    memo = [-1] * n

    def dp(index):
        if memo[index] != -1: return memo[index]
        if index == 0: return nums[0]
        if index < 0: return 0

        take = nums[index] + dp(index - 2)
        nontake = dp(index - 1)

        memo[index] = max(take, nontake)
        return memo[index]
    
    return dp(n - 1)

def max_sum_tabulation(nums: list[int]):
    n = len(nums)
    dp = [0] * n
    dp[0] = nums[0]

    for i in range(1, n):
        take = nums[i] + (dp[i - 2] if i > 1 else 0)
        nontake = dp[i - 1]
        dp[i] = max(take, nontake)

    return dp[n - 1]

def max_sum_spaceopt(nums: list[int]):
    n = len(nums)
    prev = nums[0]
    prev2 = 0

    for i in range(1, n):
        take = nums[i]
        if i > 1: take += prev2

        nontake = prev
        curr = max(take, nontake)
        prev2 = prev
        prev = curr
    
    return prev

max_sum_spaceopt([2, 1, 4, 9])
```

### House Robber - II

```python
def house_robber_2(nums: list[int]):
    n = len(nums)
    if n == 1: return nums[0]
    return max(max_sum_spaceopt(nums[:n-1]), max_sum_spaceopt(nums[1:]))

house_robber_2([2, 3, 2])
```

### Climbing Stairs II

You are climbing a staircase with n + 1 steps, numbered from 0 to n.

You are also given a 1-indexed integer array costs of length n, where costs[i] is the cost of step i.

From step i, you can jump only to step i + 1, i + 2, or i + 3. The cost of jumping from step i to step j is defined as: costs[j] + (j - i)2

You start from step 0 with cost = 0.

Return the minimum total cost to reach step n.

```python
def climbStairs(self, n, costs):
    dp = [0] * (n + 1)

    for i in range(1, n+1):
        min_energy = float('inf')
        for j in range(1, 3+1):
            if i - j >= 0:
                # i - j is the launch step
                # i is destination step, i - 1 is cost as its 1-indexed, 
                # j is jump distance - 1, 2, 3
                energy_i = dp[i - j] + costs[i - 1] + j*j
                min_energy = min(min_energy, energy_i)

        dp[i] = min_energy
    
    return dp[n]
```

### Ninja Training

```python
def ninjaTraining(n, points):
    # dp[day][last] = max points achievable from day 0..day,
    # given that task `last` cannot be performed on day+1
    # last ∈ {0, 1, 2}  -> that task is forbidden
    # last == 3         -> no restriction (all 3 tasks allowed)
    dp = [[0] * 4 for _ in range(n)]

    # ---------- Base case: Day 0 ----------
    # If `last` task is forbidden today, pick best among the other two.
    dp[0][0] = max(points[0][1], points[0][2])              # can't do 0
    dp[0][1] = max(points[0][0], points[0][2])              # can't do 1
    dp[0][2] = max(points[0][0], points[0][1])              # can't do 2
    dp[0][3] = max(points[0][0], points[0][1], points[0][2])  # no restriction

    # ---------- Fill table from Day 1 onwards ----------
    for day in range(1, n):
        for last in range(4):
            best = 0
            for task in range(3):           # try each of the 3 tasks today
                if task != last:            # skip the forbidden one
                    # Today's points + best from previous day
                    # given that today's `task` is forbidden tomorrow's view
                    earned = points[day][task] + dp[day - 1][task]
                    best = max(best, earned)
            dp[day][last] = best

    # Final answer: on the last day, no task is forbidden -> use state 3
    return dp[n - 1][3]
```
