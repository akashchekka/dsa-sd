```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
node = Node(1)
node.next = Node(2)
node.next.next = Node(3)
node.next.next.next = Node(4)
node.next.next.next.next = Node(5)

node.next.next.next.next.next = node.next.next

slow = node
fast = node

isloop = False

while True:
    try:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            isloop = True
            break
    except:
        isloop = False
        break

print(isloop)

### Below piece of code will give the point of intersection

# slow = node

# while True:
#     slow = slow.next
#     fast = fast.next
#     if slow == fast:
#         print(fast.data)
#         break
```
