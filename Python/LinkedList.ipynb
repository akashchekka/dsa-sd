```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
node = Node(1)
node.next = Node(2)
node.next.next = Node(3)
node.next.next.next = Node(4)
node.next.next.next.next = Node(5)

node.next.next.next.next.next = node.next.next

slow = node
fast = node

isloop = False

while True:
    try:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            isloop = True
            break
    except:
        isloop = False
        break

print(isloop)

### Below piece of code will give the point of intersection

# slow = node

# while True:
#     slow = slow.next
#     fast = fast.next
#     if slow == fast:
#         print(fast.data)
#         break
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

node = Node(1)
node.next = Node(3)
node.next.next = Node(5)
node2 = Node(4)
node2.next = Node(6)

def add_node(head: Node, node: Node):
    if not head:
        return node
    
    temp = head
    while(temp.next != None):
        temp = temp.next
    
    temp.next = node
    return head

temp1 = node
temp2 = node2
res = None

while(temp1 and temp2):
    if temp1.data <= temp2.data:
        res = add_node(res, Node(temp1.data))
        temp1 = temp1.next
    else:
        res = add_node(res, Node(temp2.data))
        temp2 = temp2.next

while temp1:
    res = add_node(res, Node(temp1.data))
    temp1 = temp1.next

while temp2:
    res = add_node(res, Node(temp2.data))
    temp2 = temp2.next

temp3 = res
while(temp3):
    print(temp3.data, end='=>')
    temp3 = temp3.next
print('Null')
```
