```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

node = Node(1)
node.next = Node(3)
node.next.next = Node(5)
node2 = Node(2)
node2.next = Node(4)

def add_node(head, node):
    if not head:
        return node
    
    temp = head
    while(temp.next != None):
        temp = temp.next
    
    temp.next = node
    return head

temp1 = node
temp2 = node2
res = None

while(temp1 and temp2):
    if temp1.data <= temp2.data:
        res = add_node(res, temp1)
        res = add_node(res, temp2)
    else:
        res = add_node(res, temp2)
        res = add_node(res, temp1)
    temp1 = temp1.next
    temp2 = temp2.next

while temp1:
    res = add_node(res, temp1)
    temp1 = temp1.next

while temp2:
    res = add_node(res, temp2)
    temp2 = temp2.next

temp3 = res
while(temp3):
    print(temp3.data, end='=>')
    temp3 = temp3.next
```
