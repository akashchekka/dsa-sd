### Move zeros

```python
a = [1, 2, 0, 6, 0, 3, 5]

i = 0

for j in range(len(a)):
    if a[i] == 0 and a[j] != 0:
        a[i], a[j] = a[j], a[i]
    if a[i] != 0:
        i += 1

print(a)
```

```python
b = "2[a3[2[ab]]]"
stack = []

for i in b:
    if i != ']':
        stack.append(i)
    else:
        sub_str = ''
        while stack[-1] != '[':
            sub_str = stack.pop() + sub_str
        stack.pop()

        multiplier = ''
        while stack and stack[-1].isdigit():
            multiplier = stack.pop() + multiplier

        stack.append(int(multiplier) * sub_str)

print(''.join(stack))
```

### Pascal Triangle

```python
res = [[1]]

n = 5

for i in range(n-1):
    sub_res = [1]

    for j in range(len(res[-1])):
        if j != 0:
            sub_res.append(res[-1][j] + res[-1][j-1])
    sub_res.append(1)
    
    res.append(sub_res)

res
```

```python
class Solution(object):
    def hasSameDigits(self, s: str):
        """
        :type s: str
        :rtype: bool
        """
        if len(s) > 10**5:
            return False
        
        # Convert string to list of integers for faster operations
        digits = [int(ch) for ch in s]
        
        # Keep reducing until we have 2 digits
        while len(digits) > 2:
            new_digits = []
            for i in range(len(digits) - 1):
                new_digits.append((digits[i] + digits[i+1]) % 10)
            digits = new_digits
        
        # Check if final two digits are the same
        return digits[0] == digits[1]
    
    def hasSameDigitsII(self, s):
        """
        :type s: str
        :rtype: bool
        """
        if len(s) == 2:
            return s[0] == s[1]
        
        else:
            sub_str = ''
            for i in range(len(s) - 1):
                sub_str += str((int(s[i]) + int(s[i+1]))%10)

            return self.hasSameDigits(sub_str)
    
s = Solution()
s.hasSameDigits('8712638172')
```

### Duplicate in list without altering array

```python
nums = [1, 3, 4, 2, 5, 6, 4]

def duplicate_number(nums: list[int]):
    slow, fast = 0, 0

    while True:
        slow = nums[slow]
        fast = nums[nums[fast]]
        if slow == fast:
            break

    slow2 = 0
    while True:
        slow = nums[slow]
        slow2 = nums[slow2]

        if slow == slow2:
            return slow

print(duplicate_number(nums))        
```

### Merge Intervals

```python
intervals = [[1, 3], [8, 10], [2, 6], [12, 14]]

intervals.sort(key= lambda x : x[0])
merged = [intervals[0]]

for i in range(1, len(intervals)):
    current_interval = intervals[i]
    merged_interval = merged[-1]

    if (current_interval[0] <= merged_interval[1]):
        merged_interval[1] = max(merged_interval[1], current_interval[1])
    else:
        merged.append(current_interval)

merged
```

### Kadane's Algorithm

```python
nums = [-1, 1, -3, 2, -1, 4, 1]

def max_sum_subarray(nums):
    currentSum = 0
    maxSum = 0

    for i in nums:
        currentSum = max(currentSum, 0)
        currentSum += i
        maxSum = max(maxSum, currentSum)

    return maxSum

max_sum_subarray(nums)
```

### Longest consecutive chars

```python
char = 'abcdeftuvwxyz'

def chars_to_ints(string):
    return [ord(c.lower()) - ord('a') for c in string if c.isalpha()]

char_array = chars_to_ints(char)
print(char_array)
char_set = set(char_array)

res = 0
for i in char_array:
    if i-1 not in char_set:
        current_num = i
        current_length = 1
        while current_num + 1 in char_set:
            current_length += 1
            current_num +=1
        res = max(current_length, res)

res
```

### Group Anagrams

```python
from collections import defaultdict

arr = ['act', 'cat', 'ant', 'tan', 'nat']
hashmap = defaultdict(list)

for i in arr:
    counter = [0] * 26

    for j in i:
        counter[ord(j) - ord('a')] += 1
    
    hashmap[tuple(counter)].append(i)

list(hashmap.values())
```
