```python
class Solution(object):
    def hasSameDigits(self, s: str):
        """
        :type s: str
        :rtype: bool
        """
        if len(s) > 10**5:
            return False
        
        # Convert string to list of integers for faster operations
        digits = [int(ch) for ch in s]
        
        # Keep reducing until we have 2 digits
        while len(digits) > 2:
            new_digits = []
            for i in range(len(digits) - 1):
                new_digits.append((digits[i] + digits[i+1]) % 10)
            digits = new_digits
        
        # Check if final two digits are the same
        return digits[0] == digits[1]
    
    def hasSameDigitsII(self, s):
        """
        :type s: str
        :rtype: bool
        """
        if len(s) == 2:
            return s[0] == s[1]
        
        else:
            sub_str = ''
            for i in range(len(s) - 1):
                sub_str += str((int(s[i]) + int(s[i+1]))%10)

            return self.hasSameDigits(sub_str)
    
s = Solution()
s.hasSameDigits('8712638172')
```

### Duplicate in list without altering array

```python
nums = [1, 3, 4, 2, 5, 6, 4]

def duplicate_number(nums: list[int]):
    slow, fast = 0, 0

    while True:
        slow = nums[slow]
        fast = nums[nums[fast]]
        if slow == fast:
            break

    slow2 = 0
    while True:
        slow = nums[slow]
        slow2 = nums[slow2]

        if slow == slow2:
            return slow

print(duplicate_number(nums))        
```

### Kadane's Algorithm

```python
nums = [-1, 1, -3, 2, -1, 4, 1]

def max_sum_subarray(nums):
    currentSum = 0
    maxSum = 0

    for i in nums:
        currentSum = max(currentSum, 0)
        currentSum += i
        maxSum = max(maxSum, currentSum)

    return maxSum

max_sum_subarray(nums)
```

### Merge 2 sorted arrays

```python
nums1 = [2, 3, 4, 0, 0]
nums2 = [1, 5]

m = 0
for i in nums1:
    if i != 0:
        m += 1

n = len(nums2)

last = m + n - 1
m -= 1
n -= 1

while m >= 0 and n >= 0:
    if nums1[m] < nums2[n]:
        nums1[last] = nums2[n]
        n -= 1
    else:
        nums1[last] = nums1[m]
        m -= 1
    last -= 1

while n >= 0:
    nums1[last] = nums2[n]
    n, last = n - 1, last - 1

nums1
```

### Product of array except self

```python
nums = [2, 3, 4, 5]
res = [1]

for i in range(1, len(nums)):
    res.append(res[i-1] * nums[i-1])
# 1, 2, 2*3*5, 2*3*4

c = len(res) - 1 - 1 # 4 - 1 - 1 = 2

postfix = nums[len(nums) - 1] # 5

while c >= 0:
    res[c] = res[c] * postfix
    postfix *= nums[c]
    c -= 1

res
```

### Max Consecutive Ones

```python
def max_ones(nums: list[int], k: int):
    l = res = zeros = 0

    for r in range(len(nums)):
        if nums[r] == 0:
            zeros += 1

        if zeros > k:
            if nums[l] == 0:
                zeros -= 1
            l += 1

        res = max(res, r - l + 1)

    return res

max_ones([1,1,0,0,1,1,1,0,1,1], 2)
```

### Longest character replacement

```python
def characterReplacement(s: str, k: int) -> int:
    l = maxLen = maxF = 0
    c = defaultdict(int)

    for r in range(len(s)):
        c[s[r]] += 1
        maxF = max(maxF, c[s[r]])

        if (r - l + 1) - maxF > k:
            c[s[l]] -= 1
            l += 1
        else:
            maxLen = max(maxLen, r - l + 1)
    
    return maxLen
```

### Police and Thief

```python
from collections import deque

def police_and_thief(arr: list, k: int):
    p, t = deque(), deque()
    res = 0

    for i in range(len(arr)):
        if arr[i] == 'P':
            p.append(i)
        else:
            t.append(i)

    while p and t:
        if abs(p[0] - t[0]) <= k:
            res += 1
            p.popleft()
            t.popleft()
        elif p[0] < t[0]:
            p.popleft()
        else:
            t.popleft()

    return res

police_and_thief(['P', 'P', 'P', 'T', 'T'], 2)
```

### Max Product SubArray

```python
def max_product_subarray(nums: list[int]):
    pre = suf = 1
    n = len(nums)
    res = float('-inf')

    for i in range(n):
        if pre == 0: pre = 1
        if suf == 0: suf = 1

        pre = pre * nums[i]
        suf = suf * nums[n - i - 1]

        res = max(res, pre, suf)

    return res

max_product_subarray([2, 3, -1, 4])
```

### Maximum Sliding Window

```python
from collections import deque

def max_sliding_window(nums: list[int], k: int):
    n = len(nums)
    res = []

    for i in range(n - k + 1):
        res.append(max(nums[i:i + k]))

    return res

def max_sliding_window_2(nums: list[int], k: int):
    q = deque()
    n = len(nums)
    res = []
    
    for i in range(n):
        if q and q[0] <= i - k:
            q.popleft()

        while q and nums[q[-1]] < nums[i]:
            q.pop()

        q.append(i)

        if i >= k - 1:
            res.append(nums[q[0]])

    return res

print(max_sliding_window_2([1, 2, 3, 4, 5, 6], 3))
```

### Next Greater Element

```python
def nextGreaterElements(nums):
        n = len(nums)
        res = [0] * n
        stack = []

        for i in reversed(range(2*n - 1)):
            while stack and stack[-1] <= nums[i%n]:
                stack.pop()

            res[i%n] = stack[-1] if stack else -1

            stack.append(nums[i%n])

        return res
```

### Trapping Rainwater

```python
def trapping_rain_water(nums):
    l, r = 0, len(nums) - 1
    maxl, maxr = nums[l], nums[r]
    res = 0

    while l < r:
        if maxl < maxr:
            l += 1
            maxl = max(maxl, nums[l])
            res += maxl - nums[l]
        else:
            r -= 1
            maxr = max(maxr, nums[r])
            res += maxr - nums[r]

    return res

trapping_rain_water([0,2,0,3,1,0,1,3,2,1])
```

### Rearrange alternative numbers in place

```python
def rearrange_nums(nums: list[int]):
    def rotate_right(nums, start, end):
        temp = nums[end]
        for i in range(end, start, -1):
            nums[i] = nums[i - 1]
        nums[start] = temp
    
    out_of_place = -1
    for i in range(len(nums)):
        if out_of_place == -1:
            if (i%2 == 0) == (nums[i] < 0):
                out_of_place = i
        else:
            if (nums[i] < 0) == (nums[out_of_place] >= 0):
                rotate_right(nums, out_of_place, i)

                if i - out_of_place >= 2:
                    out_of_place += 2
                else:
                    out_of_place = -1
    
    return nums

rearrange_nums([1, 2, 3, -4, -1, 4])    
```
