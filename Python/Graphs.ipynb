### Has Path

```python
def hasPath(graph, src, dst):
    stack = [src]

    while stack:
        current = stack.pop()
        for neighbor in graph[current]:
            if neighbor == dst:
                return True
            stack.append(neighbor)
    
    return False

def hasPathRec(graph, src, dst):
    if (src == dst): return True

    for neighbor in graph[src]:
        if hasPathRec(graph, neighbor, dst):
            return True
    
    return False
```

### Connected Components

```python
def connectedComponents(graph):
    visited = set()
    count = 0

    def dfs(node):
        visited.add(node)

        for nei in graph[node]:
            if nei not in visited:
                dfs(nei)

    for node in graph:
        if node not in visited:
            dfs(node)
            count += 1

    print(count)

graph = {
    0: [6, 1, 5],
    1: [0],
    5: [0, 6],
    6: [0, 5],
    2: [3, 4],
    3: [2, 4],
    4: [3, 2]
}

connectedComponents(graph=graph)
```

### Largest Component

```python
def largestComponent(graph):
    visited = set()
    count = 0

    def explore(node):
        if node in visited: return 0

        visited.add(node)
        size = 1
        for nei in graph[node]:
            size += explore(nei)

        return size
    
    for node in graph:
        # if node not in visited:
        count = max(count, explore(node))

    print(count)

graph = {
    0: [6, 1, 5],
    1: [0],
    5: [0, 6],
    6: [0, 5],
    2: [3, 4],
    3: [2, 4],
    4: [3, 2]
}

largestComponent(graph=graph)
```

### Build Adjacency Graph

```python
def buildGraph(edges):
    graph = dict()

    for edge in edges:
        if edge[0] not in graph:
            graph[edge[0]] = []
        if edge[1] not in graph:
            graph[edge[1]] = []

        graph[edge[0]].append(edge[1])
        graph[edge[1]].append(edge[0])

    return graph
```

### Shortest Path in Undirected graph

```python
from collections import deque

def shortest_path(graph, nodeA, nodeB):
    visited = set()
    q = deque()
    q.append([nodeA, 0])
    visited.add(nodeA)
    
    while q:
        [current, distance] = q.popleft()
        if current == nodeB: return distance

        for nei in graph[current]:
            if nei not in visited:
                visited.add(nei)
                q.append([nei, distance + 1])
    
    return -1

print(shortest_path(buildGraph([
    ['w', 'x'],
    ['x', 'y'],
    ['z', 'y'],
    ['z', 'v'],
    ['w', 'v']
]), 'w', 'z'))
```

### Clone Graph

```python
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

def clone_graph(node):
    if not node:
        return None
    
    newGraph = dict()

    def dfs(node):
        if node in newGraph:
            return newGraph[node]
        
        clone = Node(node.val)
        newGraph[node] = clone

        for nei in node.neighbors:
            clone.neighbors.append(dfs(nei))

        return clone
    
    return dfs(node)

graph = {
    0: [6, 1, 5],
    1: [0],
    5: [0, 6],
    6: [0, 5],
    2: [3, 4],
    3: [2, 4],
    4: [3, 2]
}

print(clone_graph(graph))
```

### Detect Cycle

```python
def has_cycle(graph):
    visited = set()

    def explore(node, parent):
        visited.add(node)

        for nei in graph[node]:
            if nei == parent:
                continue
            if nei in visited:
                return True
            if explore(nei, node):
                return True
        
        return False
    
    for node in graph:
        if node not in visited:
            if explore(node, -1):
                return True
            
    return False
```

```python
print(has_cycle(buildGraph([[0, 1], [1, 2], [2, 3]])))
```

### Topological sort

```python
from collections import deque, defaultdict

def topological_sort(edges: list[list[str]]):
    in_degree = defaultdict(int)
    graph = defaultdict(list)

    nodes = set()
    for u, v in edges:
        graph[u].append(v)
        in_degree[v] += 1
        nodes.add(u)
        nodes.add(v)

    q = deque([node for node in nodes if in_degree[node] == 0])

    res = []

    while q:
        current = q.popleft()
        res.append(current)

        for nei in graph[current]:
            in_degree[nei] -= 1
            if in_degree[nei] == 0:
                q.append(nei)

    return res

res = topological_sort([
    ['A', 'C'],
    ['B', 'C'],
    ['C', 'E'],
    ['B', 'D'],
    ['E', 'F'],
    ['D', 'F']
])

res
```
