### Has Path

```python
def hasPath(graph, src, dst):
    stack = [src]

    while stack:
        current = stack.pop()
        for neighbor in graph[current]:
            if neighbor == dst:
                return True
            stack.append(neighbor)
    
    return False

def hasPathRec(graph, src, dst):
    if (src == dst): return True

    for neighbor in graph[src]:
        if hasPathRec(graph, neighbor, dst):
            return True
    
    return False
```

### Connected Components

```python
def connectedComponents(graph):
    visited = set()
    count = 0

    def dfs(node):
        visited.add(node)

        for nei in graph[node]:
            if nei not in visited:
                dfs(nei)

    for node in graph:
        if node not in visited:
            dfs(node)
            count += 1

    print(count)

graph = {
    0: [6, 1, 5],
    1: [0],
    5: [0, 6],
    6: [0, 5],
    2: [3, 4],
    3: [2, 4],
    4: [3, 2]
}

connectedComponents(graph=graph)
```

### Largest Component

```python
def largestComponent(graph):
    visited = set()
    count = 0

    def explore(node):
        if node in visited: return 0

        visited.add(node)
        size = 1
        for nei in graph[node]:
            size += explore(nei)

        return size
    
    for node in graph:
        # if node not in visited:
        count = max(count, explore(node))

    print(count)

graph = {
    0: [6, 1, 5],
    1: [0],
    5: [0, 6],
    6: [0, 5],
    2: [3, 4],
    3: [2, 4],
    4: [3, 2]
}

largestComponent(graph=graph)
```

### Build Adjacency Graph

```python
def buildGraph(edges):
    graph = dict()

    for edge in edges:
        if edge[0] not in graph:
            graph[edge[0]] = []
        if edge[1] not in graph:
            graph[edge[1]] = []

        graph[edge[0]].append(edge[1])
        graph[edge[1]].append(edge[0])

    return graph
```

### Shortest Path in Undirected graph

```python
from collections import deque

def shortest_path(graph, nodeA, nodeB):
    visited = set()
    q = deque()
    q.append([nodeA, 0])
    visited.add(nodeA)
    
    while q:
        [current, distance] = q.popleft()
        if current == nodeB: return distance

        for nei in graph[current]:
            if nei not in visited:
                visited.add(nei)
                q.append([nei, distance + 1])
    
    return -1

print(shortest_path(buildGraph([
    ['w', 'x'],
    ['x', 'y'],
    ['z', 'y'],
    ['z', 'v'],
    ['w', 'v']
]), 'w', 'z'))
```

### Clone Graph

```python
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

def clone_graph(node):
    if not node:
        return None
    
    mapping = dict()

    def dfs(node):
        if node in mapping:
            return mapping[node]
        
        clone = Node(node.val)
        mapping[node] = clone

        for nei in node.neighbors:
            clone.neighbors.append(dfs(nei))

        return clone
    
    return dfs(node)

graph = {
    0: [6, 1, 5],
    1: [0],
    5: [0, 6],
    6: [0, 5],
    2: [3, 4],
    3: [2, 4],
    4: [3, 2]
}

print(clone_graph(graph))
```

### Detect Cycle

```python
def has_cycle(graph):
    visited = set()

    def explore(node, parent):
        visited.add(node)

        for nei in graph[node]:
            if nei == parent:
                continue
            if nei in visited:
                return True
            if explore(nei, node):
                return True
        
        return False
    
    for node in graph:
        if node not in visited:
            if explore(node, -1):
                return True
            
    return False

print(has_cycle(buildGraph([[0, 1], [1, 2], [2, 3]])))
```

### Topological sort

```python
from collections import deque, defaultdict
from heapq import heapify, heappush, heappop

def topological_sort(edges: list[list[str]]):
    in_degree = defaultdict(int)
    graph = defaultdict(list)

    nodes = set()
    for u, v in edges:
        graph[u].append(v)
        in_degree[v] += 1
        nodes.add(u)
        nodes.add(v)

    q = deque([node for node in nodes if in_degree[node] == 0])

    res = []

    while q:
        current = q.popleft()
        res.append(current)

        for nei in graph[current]:
            in_degree[nei] -= 1
            if in_degree[nei] == 0:
                q.append(nei)

    return res

def topo_sort_heap(edges: list[list[str]]):
    indegree = defaultdict(int)
    graph = defaultdict(list)

    nodes = set()
    for u, v in edges:
        graph[u].append(v)
        indegree[v] += 1
        nodes.add(u)
        nodes.add(v)
    
    heap = [node for node in nodes if indegree[node] == 0]
    heapify(heap)
    res = []

    while heap:
        curr = heappop(heap)
        res.append(curr)

        for nei in graph[curr]:
            indegree[nei] -= 1
            if indegree[nei] == 0:
                heappush(heap, nei)

    return res

res = topological_sort([
    ['A', 'C'],
    ['B', 'C'],
    ['C', 'E'],
    ['B', 'D'],
    ['E', 'F'],
    ['D', 'F']
])

res
```

### Alien Dictionary

```python
from collections import deque

def alien_dict(words: list[list[str]]):
    graph = {node: set() for word in words for node in word}
    indegree = {node: 0 for node in graph}

    for i in range(len(words) - 1):
        w1 = words[i]
        w2 = words[i + 1]

        minlen = min(len(w1), len(w2))
        if len(w1) > len(w2) and w1[:minlen] == w2[:minlen]:
            return ""

        for j in range(minlen):
            if w1[j] != w2[j]:
                if w2[j] not in graph[w1[j]]:
                    graph[w1[j]].add(w2[j])
                    indegree[w2[j]] += 1
                break

    q = deque([node for node in indegree if indegree[node] == 0])
    res = []
    while q:
        current = q.popleft()
        res.append(current)

        for nei in graph[current]:
            indegree[nei] -= 1
            if indegree[nei] == 0:
                q.append(nei)

    return ''.join(res) if len(res) == len(indegree) else ""

words = ["baa", "abcd", "abca", "cab", "cad"]
alien_dict(words)
```

### Word Ladder

```python
from collections import defaultdict, deque

def word_ladder(begin: str, end: str, wordlist: list[str]):
    if end not in wordlist:
        return 0
    
    nei = defaultdict(list)
    wordlist.append(begin)

    for word in wordlist:
        for j in range(len(word)):
            pattern = word[:j] + '*' + word[j + 1:]
            nei[pattern].append(word)

    visited = set([begin])
    q = deque([begin])
    res = 1
    while q:
        for _ in range(len(q)):
            word = q.popleft()
            if word == end:
                return res
            for j in range(len(word)):
                pattern = word[:j] + '*' + word[j + 1:]
                for neiWord in nei[pattern]:
                    if neiWord not in visited:
                        visited.add(neiWord)
                        q.append(neiWord)

        res += 1
    return 0

word_ladder('cat', 'sag', ["bat","bag","dag","dot", "sag"])
```

### Bus Routes

```python
from collections import defaultdict, deque

def numBusesToDestination(routes, source, target):
    if source == target:
        return 0

    # Map: stop → list of bus indices that visit this stop
    stop_to_buses = defaultdict(set)
    for bus, stops in enumerate(routes):
        for stop in stops:
            stop_to_buses[stop].add(bus)

    # BFS
    visited_buses = set()
    visited_stops = set()
    visited_stops.add(source)
    q = deque()
    q.append(source)
    buses_taken = 0 

    while q:
        buses_taken += 1
        for _ in range(len(q)):  # process level by level
            stop = q.popleft()
            for bus in stop_to_buses[stop]:
                if bus in visited_buses:
                    continue
                visited_buses.add(bus)
                for next_stop in routes[bus]:
                    if next_stop == target:
                        return buses_taken
                    if next_stop not in visited_stops:
                        visited_stops.add(next_stop)
                        q.append(next_stop)

    return -1

numBusesToDestination([[6,7],[3,6,7]], 1, 6)
```

### Dijkstra's Algorithm

```python
from heapq import heappush, heappop

def dijkstra_algo(adj: dict[tuple[int, int]], n: int):
    dis = [float('inf')] * n
    
    pq = []
    dis[0] = 0
    heappush(pq, (0, 0))

    while pq:
        current_dis, current_node = heappop(pq)

        if current_dis > dis[current_node]:
            continue

        for nei_dis, nei_node in adj[current_node]:
            if current_dis + nei_dis < dis[nei_node]:
                dis[nei_node] = current_dis + nei_dis
                heappush(pq, (dis[nei_node], nei_node))

    return dis

n = 5
adj = {
    0: [(2, 1), (4, 2)],
    1: [(1, 2), (7, 3)],
    2: [(3, 4)],
    3: [(1, 4)],
    4: []
}

dijkstra_algo(adj, n)
```

### Cheapest flights with k stops

```python
from collections import deque

def cheapest_flights(flights: list[list[int]], n: int, src: int, dst: int, k: int):
    graph = [[] for _ in range(n)]
    for frm, to, price in flights:
        graph[frm].append((to, price))

    dist = [float('inf')] * n

    q = deque()
    # (stops, node, dst)
    q.append((0, src, 0))

    while q:
        stop, curr_node, curr_dist = q.popleft()

        for nei_node, nei_dst in graph[curr_node]:
            if curr_dist + nei_dst < dist[nei_node] and stop <= k:
                dist[nei_node] = curr_dist + nei_dst
                q.append((stop + 1, nei_node, dist[nei_node]))

    return -1 if dist[dst] == float('inf') else dist[dst]

cheapest_flights([[0,1,200],[1,2,100],[1,3,300],[2,3,100]], 4, 0, 3, 1)
```
