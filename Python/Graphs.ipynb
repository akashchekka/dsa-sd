### Has Path

```python
def hasPath(graph, src, dst):
    stack = [src]

    while stack:
        current = stack.pop()
        for neighbor in graph[current]:
            if neighbor == dst:
                return True
            stack.append(neighbor)
    
    return False

def hasPathRec(graph, src, dst):
    if (src == dst): return True

    for neighbor in graph[src]:
        if hasPathRec(graph, neighbor, dst):
            return True
    
    return False
```

### Connected Components

```python
def connectedComponents(graph):
    visited = set()
    count = 0

    def dfs(node):
        visited.add(node)

        for nei in graph[node]:
            if nei not in visited:
                dfs(nei)

    for node in graph:
        if node not in visited:
            dfs(node)
            count += 1

    print(count)
```

### Largest Component

```python
def largestComponent(graph):
    visited = set()
    count = 0

    def explore(node):
        if node in visited: return 0

        visited.add(node)
        size = 1
        for nei in graph[node]:
            size += explore(nei)

        return size
    
    for node in graph:
        # if node not in visited:
        count = max(count, explore(node))

    print(count)
```

### Build Adjacency Graph

```python
def buildGraph(edges):
    graph = dict()

    for edge in edges:
        if edge[0] not in graph:
            graph[edge[0]] = []
        if edge[1] not in graph:
            graph[edge[1]] = []

        graph[edge[0]].append(edge[1])
        graph[edge[1]].append(edge[0])

    return graph
```

### Shortest Path in Undirected graph

```python
from collections import deque

def shortest_path(graph, nodeA, nodeB):
    visited = set()
    q = deque()
    q.append([nodeA, 0])
    visited.add(nodeA)
    
    while q:
        [current, distance] = q.popleft()
        if current == nodeB: return distance

        for nei in graph[current]:
            if nei not in visited:
                visited.add(nei)
                q.append([nei, distance + 1])
    
    return -1
```

### Number of Islands

```python
def no_of_islands(grid: list[list[str]]):
    visited = set()
    count = 0
    rows, cols = len(grid), len(grid[0])

    def explore(r, c):
        # Check r and c boundaries
        rowInbounds = 0 <= r < rows
        colInbounds = 0 <= c < cols
        if not rowInbounds or not colInbounds: return False
        
        # Check if its water
        if grid[r][c] == '0': return False

        pos = str(r) + ',' + str(c)
        if pos in visited: return False

        visited.add(pos)

        explore(r - 1, c)
        explore(r + 1, c)
        explore(r, c - 1)
        explore(r, c + 1)

        return True

    for i in range(rows):
        for j in range(cols):
            if explore(i, j):
                count += 1

    print(count)
```

```python
edges = [
    ['i', 'j'],
    ['k', 'i'],
    ['m', 'k'],
    ['k', 'l'],
    ['o', 'n']
]

print(buildGraph(edges))
```
