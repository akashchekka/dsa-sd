### Has Path

```python
def hasPath(graph, src, dst):
    stack = [src]

    while stack:
        current = stack.pop()
        for neighbor in graph[current]:
            if neighbor == dst:
                return True
            stack.append(neighbor)
    
    return False

def hasPathRec(graph, src, dst):
    if (src == dst): return True

    for neighbor in graph[src]:
        if hasPathRec(graph, neighbor, dst):
            return True
    
    return False
```

### Connected Components

```python
def connectedComponents(graph):
    visited = set()
    count = 0

    def dfs(node):
        visited.add(node)

        for nei in graph[node]:
            if nei not in visited:
                dfs(nei)

    for node in graph:
        if node not in visited:
            dfs(node)
            count += 1

    print(count)
```

### Largest Component

```python
def largestComponent(graph):
    visited = set()
    count = 0

    def explore(node):
        if node in visited: return 0

        visited.add(node)
        size = 1
        for nei in graph[node]:
            size += explore(nei)

        return size
    
    for node in graph:
        # if node not in visited:
        count = max(count, explore(node))

    print(count)
```

### Build Adjacency Graph

```python
def buildGraph(edges):
    graph = dict()

    for edge in edges:
        if edge[0] not in graph:
            graph[edge[0]] = []
        if edge[1] not in graph:
            graph[edge[1]] = []

        graph[edge[0]].append(edge[1])
        graph[edge[1]].append(edge[0])

    return graph
```

### Shortest Path in Undirected graph

```python
from collections import deque

def shortest_path(graph, nodeA, nodeB):
    visited = set()
    q = deque()
    q.append([nodeA, 0])
    visited.add(nodeA)
    
    while q:
        [current, distance] = q.popleft()
        if current == nodeB: return distance

        for nei in graph[current]:
            if nei not in visited:
                visited.add(nei)
                q.append([nei, distance + 1])
    
    return -1
```

### Clone Graph

```python
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

def clone_graph(node):
    if not node:
        return None
    
    newGraph = dict()

    def dfs(node):
        if node in newGraph:
            return newGraph[node]
        
        clone = Node(node.val)
        newGraph[node] = clone

        for nei in node.neighbors:
            clone.neighbors.append(dfs(nei))

        return clone
    
    return dfs(node)
```

### Number of Islands

```python
def no_of_islands(grid: list[list[str]]):
    visited = set()
    count = 0
    rows, cols = len(grid), len(grid[0])

    def explore(r, c):
        if ((not (0 <= r < rows) or not (0 <= c < cols)) or
            grid[r][c] == "0" or
            (r, c) in visited): return False

        visited.add((r, c))

        explore(r - 1, c)
        explore(r + 1, c)
        explore(r, c - 1)
        explore(r, c + 1)

        return True

    for i in range(rows):
        for j in range(cols):
            if explore(i, j):
                count += 1

    print(count)
```

### Max Area of Island

```python
def max_area_of_islands(grid: list[list[str]]):
    visited = set()
    count = 0
    rows, cols = len(grid), len(grid[0])

    def explore(r, c):
        if ((not (0 <= r < rows) or not (0 <= c < cols)) or
            grid[r][c] == 0 or
            (r, c) in visited): return 0

        visited.add((r, c))

        size = 1 + explore(r - 1, c) + explore(r + 1, c) + explore(r, c - 1) + explore(r, c + 1)

        return size

    for i in range(rows):
        for j in range(cols):
            count = max(count, explore(i, j))

    print(count)
```

### Rotten Oranges

```python
from collections import deque

def rotten_oranges(grid: list[list[int]]):
    q = deque()
    fresh = 0
    rows, cols = len(grid), len(grid[0])

    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == 1:
                fresh += 1
            if grid[r][c] == 2:
                q.append((r, c))

    minutes = 0
    while q and fresh:
        minutes += 1

        for _ in range(len(q)):
            iters = [(1, 0), (-1, 0), (0, 1), (0, -1)]
            r, c = q.popleft()
            for dr, dc in iters:
                if ((0 <= r + dr < rows) and (0 <= c + dc < cols) and  grid[r + dr][c + dc] == 1):
                    grid[r + dr][c + dc] = 2
                    q.append((r + dr, c + dc))
                    fresh -= 1

    return minutes if fresh == 0 else -1
```

### Surrounded Region

```python
def surrounded_region(grid: list[list[str]]):
    rows, cols = len(grid), len(grid[0])

    def capture(r, c):
        if (r < 0 or r == rows or c < 0 or c == cols or grid[r][c] != 'O'):
            return
        
        grid[r][c] = 'T'
        capture(r + 1, c)
        capture(r - 1, c)
        capture(r, c + 1)
        capture(r, c - 1)

    for r in range(rows):
        for c in range(cols):
            if (grid[r][c] == 'O' and 
                (r in [0, rows - 1] or c in [0, cols - 1])):
                capture(r, c)

    for r in range(rows):
        for c in range(cols):
            if (grid[r][c] == 'O'):
                grid[r][c] = 'X'

    for r in range(rows):
        for c in range(cols):
            if (grid[r][c] == 'T'):
                grid[r][c] = 'O'

    return grid
```

### Word Search

```python
def word_search(grid: list[list[str]], word: str):
    rows, cols = len(grid), len(grid[0])
    visited = set()

    def dfs(r, c, index):
        if index == len(word): return True

        if (not (0 <= r < rows) or 
            not (0 <= c < cols) or 
             grid[r][c] != word[index] or
             (r, c) in visited) : return False

        visited.add((r, c))
        isValid = False

        for dr, dc in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
            nr, nc = r + dr, c + dc
            isValid = isValid or bfs(nr, nc, index + 1)

        visited.remove((r, c))
        return isValid

    for i in range(rows):
        for j in range(cols):
            if dfs(i, j, 0):
                return True
    
    return False
```

### Islands and Tresure

```python
from collections import deque

def islandsAndTresure(grid: list[list[int]]):
    rows, cols = len(grid), len(grid[0])
    q = deque()
    visited = set()

    def addland(r, c):
        if(not (0 <= r < rows) or
            not (0 <= c < cols) or
             (r, c) in visited or
              grid[r][c] == -1):
            return
        
        q.append((r, c))
        visited.add((r, c))

    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == 0:
                q.append((r, c))
                visited.add((r, c))

    dist = 0
    while q:
        for _ in range(len(q)):
            r, c = q.popleft()
            grid[r][c] = dist
            addland(r + 1, c)
            addland(r - 1, c)
            addland(r, c + 1)
            addland(r, c - 1)
        dist += 1

    print(grid)
```
