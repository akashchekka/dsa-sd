### Merge K Lists

```python
def merge_k_lists(lists):
    if not lists or len(lists) == 0:
        return None
    
    def merge_lists(l1, l2):
        dummy = Node()
        temp = dummy

        while l1 and l2:
            if l1.data < l2.data:
                temp.next = l1
                l1 = l1.next
            else:
                temp.next = l2
                l2 = l2.next
            temp = temp.next

        if l1:
            temp.next = l1

        if l2:
            temp.next = l2

        return dummy.next

    while len(lists) > 1:
        merged_lists = []

        for i in range(0, len(lists), 2):
            l1 = lists[i]
            l2 = lists[i + 1] if (i + 1) < len(lists) else None

            merged_lists.append(merge_lists(l1, l2))

        lists = merged_lists

    return lists[0]
```

### Reorder list

```python
slow = node
fast = node

while fast.next and fast.next.next:
    slow = slow.next
    fast = fast.next.next

second = slow.next
slow.next = None

stack = []
while second:
    stack.append(second)
    second = second.next

temp = node

while stack:
    current = stack.pop()
    current.next = temp.next
    temp.next = current
    temp = temp.next.next
```
